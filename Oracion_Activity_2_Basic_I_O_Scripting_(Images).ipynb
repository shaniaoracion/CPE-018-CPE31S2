# **1. Loading the RGB .png image.**

````python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/cxv.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Original PNG Image (RGB)")
plt.show()
````

# **2. Changing the file type from png to jpg.**

````python
# Save image as JPG
cv2.imwrite('/content/character.jpg', img)

# Load the JPG image
img_jpg = cv2.imread('/content/character.jpg')
img_jpg_rgb = cv2.cvtColor(img_jpg, cv2.COLOR_BGR2RGB)

# Display JPG image
plt.imshow(img_jpg_rgb)
plt.axis('off')
plt.title("Converted JPG Image")
plt.show()
````

# **3. Create a grayscale version of the original image**

````python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display grayscale image
plt.imshow(gray, cmap='gray')
plt.axis('off')
plt.title("Grayscale Image")
plt.show()
````

# **4. Use all 6 enumerators of cv2.imread() with observations**

````python
flags = {
    "IMREAD_COLOR (1)": cv2.IMREAD_COLOR,
    "IMREAD_GRAYSCALE (0)": cv2.IMREAD_GRAYSCALE,
    "IMREAD_UNCHANGED (-1)": cv2.IMREAD_UNCHANGED,
    "IMREAD_REDUCED_COLOR_2": cv2.IMREAD_REDUCED_COLOR_2,
    "IMREAD_REDUCED_COLOR_4": cv2.IMREAD_REDUCED_COLOR_4,
    "IMREAD_REDUCED_COLOR_8": cv2.IMREAD_REDUCED_COLOR_8
}

plt.figure(figsize=(12, 8))

for i, (name, flag) in enumerate(flags.items(), 1):
    image = cv2.imread('/cxv.png', flag)
    plt.subplot(2, 3, i)
    if len(image.shape) == 2:
        plt.imshow(image, cmap='gray')
    else:
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(name)
    plt.axis('off')

plt.show()
````

IMREAD_COLOR  Loads a full-color image (default behavior).

IMREAD_GRAYSCALE  Converts the image into a single-channel grayscale image.

IMREAD_UNCHANGED  Preserves the original image including alpha channel if present.

IMREAD_REDUCED_COLOR_2  Loads color image at half resolution.

IMREAD_REDUCED_COLOR_4  Loads color image at quarter resolution.

IMREAD_REDUCED_COLOR_8  Loads color image at one-eighth resolution.

# **5. Obtain an ROI that captures the full face of the character**

````python
# Image Dimensions
h, w, _ = img.shape
print(h, w)

# Define ROI
x1, y1 = int(w*0.35), int(h*0.20)
x2, y2 = int(w*0.65), int(h*0.55)

face_roi = img[y1:y2, x1:x2]

# Display the face ROI
plt.imshow(cv2.cvtColor(face_roi, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.title("Face ROI")
plt.show()
````

# **6-7. Save the face ROI as "character_face.png"**

````python
cv2.imwrite('/content/character_face.png', face_roi)
print("character_face.png saved successfully!")
````
